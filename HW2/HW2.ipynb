```python
#u = e^-x, x = -ln(u), g(u) = dx/du = -1/u
import numpy as np
import matplotlib.pyplot as plt
fx = np.random.uniform(0,1,size=1000)
plt.hist(fx,bins=20,density=True)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('f(x) Random Uniform Distribution')
```

```python
ux = [np.exp(-x) for x in fx]
plt.hist(ux,bins=20,density=True)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('e^(-x) Transformation of f(x)')
```

```python
gu = [-1/x for x in ux]
plt.hist(gu,bins=20,density=True)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Probability Density Function of Transformation u = e^(-x)')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm

# Parameters for the distributions
N, p = 10, 0.02
lamda = .2
mu, sigma = .2, 0.45

# Create a range of x values
x = np.linspace(0, 10, 1000)

# Calculating the probability density functions
binomialfunc = binom.pmf(x, N, p)
poissonfunc = poisson.pmf(x, lamda)
normalfunc = norm.pdf(x, mu, sigma)

# Plotting the functions
plt.plot(x, binomialfunc, label='Binomial(N=10, p=0.02)')
plt.plot(x, poissonfunc, label='Poisson(λ=.2)')
plt.plot(x, normalfunc, label='Normal(μ=.2, σ=0.45)')

plt.xlabel('x')
plt.ylabel('Probability Density Function')
plt.title('Probability Distribution Functions')
plt.legend()
plt.show()
from scipy.stats import kurtosis, skew

# Calculating the numerical values
binom_mean = N * p
binom_sigma = np.sqrt(N * p * (1 - p))
binom_skewness = (1 - 2 * p) / np.sqrt(N * p * (1 - p))
binom_kurtosis = 3 + (1 - 6 * p * (1 - p)) / (N * p * (1 - p))

poisson_mean = lamda
poisson_sigma = np.sqrt(lamda)
poisson_skewness = 1 / np.sqrt(lamda)
poisson_kurtosis = 1 / lamda

normal_mean = mu
normal_sigma = sigma
normal_skewness = 0
normal_kurtosis = 3

# Creating the table
import pandas as pd

data = {'Distribution': ['Binomial', 'Poisson', 'Normal'],
        'Mean': [binom_mean, poisson_mean, normal_mean],
        'Sigma': [binom_sigma, poisson_sigma, normal_sigma],
        'Skewness': [binom_skewness, poisson_skewness, normal_skewness],
        'Kurtosis': [binom_kurtosis, poisson_kurtosis, normal_kurtosis]}

df = pd.DataFrame(data)
print(df)
```
