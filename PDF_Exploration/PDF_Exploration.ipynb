```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
np.random.seed(42)
evsample = uniform.rvs(size=1000)
plt.hist(evsample)
plt.xlabel('Value')
plt.ylabel('Occurrences')
plt.title('Histogram of 1000 Uniform Random Values')
```

```python
transsamp = [x**2 for x in evsample]
#x = sqrt(u), g(u) = dx/du = 1/2 * u^(-1/2)
gofu = [1/(2*(u ** (1/2))) for u in transsamp]
plt.hist(transsamp,bins=99,histtype='step',label='u_i distribution',density=True)
plt.scatter(transsamp,gofu,label='Analytic solution')
plt.legend()
plt.title('Histogram of 1000 Uniform Random Samples Transformed w/ u = x^2')
```

```python
#Testing of methods
transsamp = [x**2 for x in evsample]
u = evsample ** 2
print(transsamp[:5])
print(u[:5])
print(type(evsample))
print(type(u))
print(type(transsamp))
```

```python
#u = 1/x, x = 1/u, g(u) = dx/du = -1/u^2
foura = [1/x for x in evsample]
fourag = [-1 /(u**(2)) for u in foura]
plt.hist(foura,bins=99,histtype='step',label='u_i distribution',density=True)
plt.scatter(foura,fourag,label='Analytic solution')
plt.legend()
plt.title('Histogram of 1000 Uniform Random Samples Transformed w/ u = 1/x')
#Technically, the limits of integration are flipped, so it follows the trend, but just inverted
```

```python
#u = sqrt(x), x = u^2, g(u) = dx/du = 2u
fourb = [x**(1/2) for x in evsample]
fourbg = [2 * u for u in fourb]
plt.hist(fourb,bins=99,histtype='step',label='u_i distribution',density=True)
plt.scatter(fourb,fourbg,label='Analytic solution')
plt.legend()
plt.title('Histogram of 1000 Uniform Random Samples Transformed w/ u = sqrt(x)')
```

```python
#u = e^x, x = ln(u), g(u) = dx/du = 1/u
fourc = [np.exp(x) for x in evsample]
fourcg = [1/u for u in fourc]
plt.hist(fourc,bins=99,histtype='step',label='u_i distribution',density=True)
plt.scatter(fourc,fourcg,label='Analytic solution')
plt.legend()
plt.title('Histogram of 1000 Uniform Random Samples Transformed w/ u = e^x')
```

```python
#u = log(x), x = 10^u, g(u) = dx/du = 10^u * ln(10)
fourd = [np.log10(x) for x in evsample]
fourdg = [(10 ** u) * np.log(10) for u in fourd]
plt.hist(fourd,bins=99,histtype='step',label='u_i distribution',density=True)
plt.scatter(fourd,fourdg,label='Analytic solution')
plt.legend()
plt.title('Histogram of 1000 Uniform Random Samples Transformed w/ u = log(x)')
```

```python
from scipy.stats import poisson
def nrepeats(N,size=5):
    mean,std = [],[]
    for i in range(N):
        events = poisson.rvs(2.0,size=size)
        mean.append(np.mean(events))
        std.append(np.std(events))
    return mean,std
def plotter(N,size=5):
    mean,std=nrepeats(N,size)
    plt.hist(mean,bins=99,histtype='step',label=f'Mu_5 mean = {np.mean(mean):.4f},Mu_5 variance = {np.var(mean):.4f}')
    plt.xlabel('Mean')
    plt.ylabel('Frequency')
    plt.title(f'Mean distribution for N={N}')
    plt.legend()
    plt.show()
    plt.hist(std,bins=99,histtype='step',label=f'Sigma_5 mean = {np.mean(std):.4f},Sigma_5 variance = {np.var(std):.4f}')
    plt.xlabel('Standard deviation')
    plt.ylabel('Frequency')
    plt.title(f'Std distribution for N={N}')
    plt.legend()
    plt.show()
plotter(10)
plotter(100)
plotter(1000)
```

```python
#CLT4
#a) Yes, they do, as is seen by the plots. This proves the central limit theorem

#b) It seems that the variance of Mu_5 converges to 0.41 & the variance of Sigma_5 converges to 0.22
#Lambda/events -> 2/5 = 0.4, (Lambda-1)/events -> 1/5 = 0.2
```

```python
plotter(1000,100)
#The respective variances of Mu_5 & sigma_5 changed their converging value to what it was divided by 20 since we increased M events by 20x
```

```python
#Bonus
from scipy.stats import gamma
means,vars = np.empty(1000),np.empty(1000)
for i in range(1000):
    gammadist = gamma.rvs(a=2,scale=1,size=100)
    means[i] = np.mean(gammadist)
    vars[i] = np.var(gammadist)

plt.hist(means,bins=99,density=True,label=f'Mean={np.mean(means):.4f}, Std={np.std(means):.4f}')
plt.title('Gamma Distribution of Means')
plt.xlabel('Value')
plt.ylabel('Frequency Normalized')
plt.legend()
plt.show()
plt.hist(vars,bins=99,density=True,label=f'Mean={np.mean(vars):.4f}, Std={np.std(vars):.4f}')
plt.title('Gamma Distribution of Variances')
plt.xlabel('Value')
plt.ylabel('Frequency Normalized')
plt.legend()
plt.show()
```
