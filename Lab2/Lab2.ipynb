```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
np.random.seed(42)
evsample = uniform.rvs(size=1000)
plt.hist(evsample)
plt.xlabel('Value')
plt.ylabel('Occurrences')
plt.title('Histogram of 1000 Uniform Random Values')
```

```python
transsamp = [x**2 for x in evsample]
#x = sqrt(u), dx/du = 1/2 * u^(-1/2)
def xsquaredfunc(u):
    return 1/2 * u ** (-1/2)
gofu = [xsquaredfunc(x) for x in transsamp]
plt.hist(transsamp,bins=100,histtype='step',label='u_i distribution')
plt.scatter(transsamp,gofu,label='Analytic solution')
plt.legend()
```
